```python
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp as mt 
from Bio.SeqUtils import molecular_weight as mw
import matplotlib.pyplot as plt 
```

```python
seqA = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
seqB = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")

#Molecular Weight
print(f"molecular weight of seqA: {mw(seqA)}")
print(f"molecular weight of seqA: {mw(seqB)}")
```

```python
#melting temp
mtA = mt.Tm_Wallace(seqA)
mtB = mt.Tm_Wallace(seqB)

plt.bar(['MT A', 'MT B'], [mtA, mtB])
plt.show()
```

```python
#GC content, PC binus pakai GC 
gcA = GC(seqA)
gcB = GC(seqB)

#AT content tapi kurang direkomendasikan
atA = 100 - gcA
atB = 100 - gcB

print(atA)
print(atB)

#ini cara manualnya, berlaku buat GC juga
# tinggal ganti hurufnya.
def AT_Content(seqX):
    AT = float((seqX.count('A')+seqX.count('T'))/len(seqX))
    return AT

print(f"AT content in seqA:{100*AT_Content(seqA)}")
print(f"AT content in seqB:{100*AT_Content(seqB)}")
plt.bar(['GC A', 'GC B'], [gcA, gcB])
plt.title("GC content comparison between two sequence")
plt.show()

plt.bar(['AT A', 'AT B'], [atA, atB])
plt.title("AT content comparison between two sequence")
plt.show()
```
